```python
n= 10 #10 period
r00= 0.05
u=1.1
d=0.9
q=.5
face_value=100.0
```

```python
import numpy as np
def short_rate_tree(n,r00,u,d):
    tree = np.zeros((n+1,n+1))
    for i in range(n+1):
        for j in range(i+1):
            tree[i][j]=r00*u**j*d**(i-j)
    return tree
```

```python
short_rate_tree(n,r00,u,d)
```

```python
def hazard_rate_tree(a,b,n):
    tree = np.zeros((n+1,n+1))
    for i in range(n):
        for j in range(i+1):
            tree[i,j]= a*b**(j-i/2)
    return tree
```

```python
hazard_rate_tree(a,b,n)
```

```python
def defaultable_bond_price(a,b,n,r00,u,d,q,face_value,R = 0.2):
    payoff = np.ones((n+1,n+1))*face_value
    rate_tree = short_rate_tree(n,r00,u,d)
    hazard_tree = hazard_rate_tree(a,b,n)
    for i in range(n-1,-1,-1):
        for j in range(i+1):
            payoff[i,j]=(q*(1-hazard_tree[i,j])*payoff[i+1,j+1]+(1-q)*(1-hazard_tree[i,j])*payoff[i+1,j])/(1+rate_tree[i,j])+ (q*hazard_tree[i,j]*R*face_value+(1-q)*hazard_tree[i,j]*R*face_value)/(1+rate_tree[i,j])
    return payoff
```

```python
defaultable_bond_price(a,b,n,r00,u,d,q,face_value,R = 0.2)
```
